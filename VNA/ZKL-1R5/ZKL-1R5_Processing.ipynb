```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
label_font = {'fontname':'Arial', 'size':'20', 'color':'black', 'weight':'normal',
          'verticalalignment':'bottom'} 
title_font = {'fontname':'sans-serif', 'size':'24', 'color':'red', 'weight':'bold'}
legend_font = {'family':'sans-serif', 'size':'16', 'style':'normal'} 


def getData(filename):
        freqs = []
        data = []
        with open(filename, 'r') as f:
                for line in f:
                        holder = line.split('\t')
                        freqs.append(float(holder[0]))
                        data.append(float(holder[1]))
        return freqs, data


def getClosest(val, arrX, arrY):
        return np.interp(val, arrX, arrY)
```

```python
filename = 'ZKL_MagReadings.csv'

gainFileManfac = 'ZKL-1R5_Gain.txt'

freqsGainManfac, ampsGainManfac = getData(gainFileManfac)

# Read data }from CSV file
data = pd.read_csv(filename)
print(data['Frequency'])
freqs = np.asarray(data['Frequency'])/1E6
gain = np.asarray(data['S21_Magnitude'])
plt.plot(freqs, gain, label = 'VNA Data')
plt.plot(freqsGainManfac, ampsGainManfac, label = 'Manufacturer')
plt.xlabel('Frequency (MHz)', labelpad = 20, **label_font)
plt.ylabel('Magnitude (dB)', **label_font)
plt.legend(loc = 'lower left', prop = legend_font)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('Gain', **title_font)
plt.show()

plt.plot(freqs, (10**((np.abs(np.asarray(gain) - getClosest(freqs, freqsGainManfac, ampsGainManfac)))/10.) - 1)*100)
plt.xlabel('Frequency (MHz)', labelpad = 20, **label_font)
plt.ylabel('Percent Difference (%)', **label_font)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('Gain Difference', **title_font)
plt.show()


```
