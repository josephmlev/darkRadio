```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget

from matplotlib.offsetbox import AnchoredText
```

```python
f = np.linspace(1,20,100) #Hz
a = 0.25 #m
b = 0.25 #m
c = 0.25 #m



def hillFormula(f,a,b,c):
    return ((8*np.pi/3)*(a*b*c)*(f**3)/((2.99e8)**3))-((a+b+c)*(f)/(2.99e8))-1/2


def dhillFormula(f,a,b,c):
    return ((8*np.pi)*(a*b*c)*(f**2)/((2.99e8)**3))-((a+b+c)/(2.99e8))

```

```python
f = np.linspace(40e6,220e6,100) #Hz
a = 2.74 #m
b = 3.05 #m
c = 4.57 #m

plt.close('all')

# Create the main plot with the first function
fig, ax1 = plt.subplots()

anchored_text = AnchoredText((f'a = {a}m\nb = {b}m\nc = {c}m'), loc=2)

ax1.plot(f/1e6, hillFormula(f, a, b, c), color='b', label='Hill Formula')
ax1.set_xlabel('Frequency (MHz)')
ax1.set_ylabel('Hill Formula Output', color='b')

# Create the secondary y-axis and plot the second function
ax2 = ax1.twinx()

ax2.plot(f/1e6, 1/(dhillFormula(f,a,b,c)*f), color='r', label='Another Function')
ax2.set_ylabel('1/(d Hill/df * f)', color='r')
ax2.semilogy()

# Add grids and show the plot
ax1.grid()
ax1.add_artist(anchored_text)
plt.show()
```
