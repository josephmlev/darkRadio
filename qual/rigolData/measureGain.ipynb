```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
calibration = np.load('./gainCalibration_TG_n40dBm_30dBAtt.npy')
gain = np.load('./gainMeasurment_TG_n40dBm_30dBAtt.npy')
freqs = gain[0]
systemGain = gain[1] - calibration[1]
```

```python
%matplotlib widget
plt.close('all')
plt.figure()
plt.plot(freqs/ 1e6, systemGain)
plt.xlabel('Frequency (MHz)')
plt.ylabel('Power (dBm)')
```

```python
#np.save('rawGain_run1_0_hz_dBm', np.array([freqs, systemGain]))
```

```python
length = 100
systemGainMedian = np.median(systemGain.reshape((-1,length)), axis = 1)
freqsMedian = np.median(freqs.reshape((-1,length)), axis = 1)
```

```python
plt.close('all')
plt.figure()
plt.plot(freqs/1e6, systemGain)
plt.plot(freqsMedian/1e6, systemGainMedian)
```
