```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some fake noise data for illustration
np.random.seed(0)
freq_wide = np.linspace(0.3, 7, 500)  # Frequency range for the wideband spectrum (in GHz)
noise_wide = np.random.normal(0, 0.1, freq_wide.size)  # Noise level
#noise_wide[400] = 1.0  # Adding a spike outside the window for illustration

freq_baseband = np.linspace(0.01, 1, 500)  # Frequency range for the baseband spectrum (in GHz)
noise_baseband = np.random.normal(0, 0.1, freq_baseband.size)  # Only noise in the baseband spectrum

# Wideband spectrum plot (300 MHz - 14 GHz)
plt.figure()
plt.plot(freq_wide, noise_wide)
plt.xlabel("Frequency (GHz)", fontsize = 24)
plt.ylabel("Power [arb.]", fontsize = 24)
plt.title("Wideband Spectrum", fontsize = 24)
plt.yticks([])  # Remove y-axis ticks
plt.xticks([0, 7], fontsize = 20)
plt.ylim(-0.3, 0.7)

# Add a shaded red box to represent the software-controlled window
plt.axvspan(2, 2.5, color='red', alpha=0.1)  # Adjust positions as needed
plt.text(2.25, 0.4, "Software Controlled\nWindow", color="red", ha="center", fontsize = 24)
plt.tight_layout()

# Add an arrow pointing right from the window label
plt.annotate('', xy=(5, 0.35), xytext=(2.5, 0.35),
             arrowprops=dict(arrowstyle="->,head_length=1,head_width=1", color="red", lw=3))

plt.show()

# Baseband spectrum plot (10 MHz - 1 GHz)
plt.figure()
plt.plot(freq_baseband*1000/2, noise_baseband)
plt.xlabel("Frequency [MHz]" , fontsize = 24)
plt.ylabel("Power [arb.]", fontsize = 24)
plt.title("Baseband Spectrum", fontsize = 24)
plt.yticks([])  # Remove y-axis ticks
plt.xticks([0, 250, 500], fontsize = 20)
plt.tight_layout()
plt.show()
```
