```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some fake noise data for illustration
np.random.seed(0)
freq_wide = np.linspace(0.3, 14, 500)  # Frequency range for the wideband spectrum (in GHz)
noise_wide = np.random.normal(0, 0.1, freq_wide.size)  # Noise level
noise_wide[400] = 1.0  # Adding a spike outside the window for illustration

freq_baseband = np.linspace(0.01, 1, 500)  # Frequency range for the baseband spectrum (in GHz)
noise_baseband = np.random.normal(0, 0.1, freq_baseband.size)  # Only noise in the baseband spectrum

# Wideband spectrum plot (300 MHz - 14 GHz)
plt.figure()
plt.plot(freq_wide, noise_wide)
plt.xlabel("Frequency (GHz)")
plt.ylabel("Power [arb.]")
plt.title("Wideband Spectrum (300 MHz - 14 GHz)")
plt.yticks([])  # Remove y-axis ticks

# Add a shaded red box to represent the software-controlled window
plt.axvspan(5, 8, color='red', alpha=0.1)  # Adjust positions as needed
plt.text(6.5, 0.8, "Software controlled\nwindow", color="red", ha="center", fontsize = 16)
plt.tight_layout

plt.show()

# Baseband spectrum plot (10 MHz - 1 GHz)
plt.figure()
plt.plot(freq_baseband*1000/2, noise_baseband)
plt.xlabel("Frequency [MHz]")
plt.ylabel("Power [arb.]")
plt.title("Baseband Spectrum (10 MHz - 1 GHz)")
plt.yticks([])  # Remove y-axis ticks
plt.tight_layout()
plt.show()
```
