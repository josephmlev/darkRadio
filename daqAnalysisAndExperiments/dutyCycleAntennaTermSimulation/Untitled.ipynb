```python
def genFreqData(spurAmp, spurIdx, numSamples):
    spurSpec = np.zeros(numSamples)
    spurSpec[spurIdx] = spurAmp
    noiseSpec = np.random.normal(mean, std, size=numSamples) 
    return noiseSpec + spurSpec
```
