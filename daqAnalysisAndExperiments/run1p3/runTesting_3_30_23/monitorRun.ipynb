```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import h5py
%matplotlib widget

dataDir         = '/drBiggerBoy/runTest_3_30_23/plottingSpec/'

antSpec         = np.load(dataDir + 'chB_W_switch0.npy')
termSpec        = np.load(dataDir + 'chB_W_switch1.npy')
antAvgSpec      = np.load(dataDir + 'chB_avg_W_switch0.npy')
termAvgSpec     = np.load(dataDir + 'chB_avg_W_switch1.npy')

freqs           = np.linspace(0, 330, 2**23)

idxStart        = 2_000_000
idxStop         = 6_000_000

avgPower_dbm    = 10*np.log10(np.mean(termSpec[idxStart:idxStop]*1000))
avgAvgPower_dbm = 10*np.log10(np.mean(termAvgSpec[idxStart:idxStop]*1000))

dataDir         = '/drBiggerBoy/runTest_3_30_23/'
totalAcqs       = int(open(dataDir + 'database.txt', 'r'
                    ).readlines()[-1].split(',')[0].strip()) + 1 

print('datetime                                : ' , str(datetime.datetime.now()))
print(f'acqNum =                                : {totalAcqs}')
print(f'freqs terminator is averaged over       : {freqs[idxStart]} - {freqs[idxStop]} MHz')
print(f'current average terminator power        : {avgPower_dbm} dBm')
print(f'average terminator power (whole run)    : {avgAvgPower_dbm} dBm')
```
