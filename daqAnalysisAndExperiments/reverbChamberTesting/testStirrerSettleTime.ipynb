Test time it takes for sirrer to settle after moving. 

method: 
zero stirrer
take many scans quickly around a peak
fit peak at each time 
plot time series of resonance
see when it settles

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget


# Plotting format for thesis
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.rc('axes', titlesize=16, labelsize=16)
plt.rc('legend', fontsize=14)
plt.rc('figure', titlesize=18)
```

```python
# Load the CSV file into a DataFrame.
df = pd.read_csv("stirrerSettleTime_300MHz.csv")

# Convert S11, S12, and S22 to complex numbers if they are stored as strings.
if df['S11'].dtype == object:
    df['S11'] = df['S11'].apply(lambda x: complex(x))
if df['S12'].dtype == object:
    df['S12'] = df['S12'].apply(lambda x: complex(x))
if df['S22'].dtype == object:
    df['S22'] = df['S22'].apply(lambda x: complex(x))

# Compute each parameter in dB (using the absolute value)
df['S11_dB'] = 20 * np.log10(np.abs(df['S11']))
df['S12_dB'] = 20 * np.log10(np.abs(df['S12']))
df['S22_dB'] = 20 * np.log10(np.abs(df['S22']))

# Group the data by 'Sweep time'. Each group corresponds to one spectrum.
grouped = df.groupby("Sweep time")

# Plot S11 spectra
plt.figure(figsize=(10, 6))
for sweep_time, group in list(grouped)[:10]:
    group_sorted = group.sort_values("Freq_MHz")
    plt.plot(group_sorted["Freq_MHz"], group_sorted["S11_dB"], label=f"{sweep_time:.3f} s")
plt.xlabel("Frequency (MHz)")
plt.ylabel("abs(S11) [dB]")
plt.title("First 10 S11 Spectra")
plt.legend()
plt.tight_layout()
plt.show()

# Plot S21 spectra
plt.figure(figsize=(10, 6))
for sweep_time, group in list(grouped)[:10]:
    group_sorted = group.sort_values("Freq_MHz")
    plt.plot(group_sorted["Freq_MHz"], group_sorted["S12_dB"], label=f"{sweep_time:.3f} s")
plt.xlabel("Frequency (MHz)")
plt.ylabel("abs(S12) [dB]")
plt.title("First 10 S12 Spectra")
plt.legend()
plt.tight_layout()
plt.show()

# Plot S22 spectra
plt.figure(figsize=(10, 6))
for sweep_time, group in list(grouped)[:10]:
    group_sorted = group.sort_values("Freq_MHz")
    plt.plot(group_sorted["Freq_MHz"], group_sorted["S22_dB"], label=f"{sweep_time:.3f} s")
plt.xlabel("Frequency (MHz)")
plt.ylabel("abs(S22) [dB]")
plt.title("First 10 S22 Spectra")
plt.legend()
plt.tight_layout()
plt.show()
```
