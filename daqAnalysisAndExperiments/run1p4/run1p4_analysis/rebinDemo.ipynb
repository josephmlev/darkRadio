```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
len             = 10000
noise           = np.random.normal(0, 1, len)
sig             = np.zeros(len)
sig[100:110]    = 2

data            = noise + sig

rebinNum        = 10
rebinData       = [] 

for idx in range(len//rebinNum):
    avgData     = np.mean(data[idx*rebinNum:(idx+1)*rebinNum])
    rebinData.append(avgData)


plt.plot(data)
plt.figure()
plt.plot(rebinData)
```
